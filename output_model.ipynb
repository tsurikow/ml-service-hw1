### **Задание 9 (0.2 балла)**

Построим нашу первую модель!
- [ ] Обучите классическую линейную регрессию с дефолтными параметрами. Посчтитайте $R^2$ и $MSE$ для трейна и для теста.

**Примечание:**

Здесь и далее $R^2$ и $MSE$ для трейна и для теста выводите везде, где требуется обучать модели, даже если в явном виде этого не просят. Иначе непонятно, как понять, насколько успешны наши эксперименты.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, root_mean_squared_error as RMSE, mean_squared_error as MSE

regr = LinearRegression()

regr.fit(X_train, y_train)

y_pred = regr.predict(X_test)
y_train_pred = regr.predict(X_train)

print("___test___")
print("MSE: %.2f" % MSE(y_test, y_pred))
print("RMSE: %.2f" % RMSE(y_test, y_pred))
print("R^2: %.2f" % r2_score(y_test, y_pred))
print("___train___")
print("MSE: %.2f" % MSE(y_train, y_train_pred))
print("RMSE: %.2f" % RMSE(y_train, y_train_pred))
print("R^2: %.2f" % r2_score(y_train, y_train_pred))
```

В мире бизнеса очень важно давать оценку качества модели понятную бизнесу, поэтому иногда заказчики приходят с кастомными метриками. Попробуем сделать такую для нашей задачи.

**Описание метрики:**

Среди всех предсказанных цен на авто нужно посчитать долю прогнозов, отличающихся от реальных цен на эти авто не более чем на 10% (в одну или другую сторону)

**Ваша задача:**

- [ ] Реализуйте метрику `business_metric`
- [ ] Посчитайте метрику для всех обученных моделей и определеите, какаю лучше всего решает задачу бизнеса

```python
def business_metric(y_true, y_pred):
    y_true = pd.DataFrame(y_true)
    y_pred = pd.Series(y_pred)
    df_metric = pd.concat([y_true, y_pred], axis=1)
    df_metric.columns = ['true', 'pred']
    # четко не указано, от какой величины считать процент, поэтому взят % от true
    df_metric['metric'] = df_metric.apply(lambda x: 1 if 0.9 <= x['pred']/x['true'] <= 1.1 else 0, axis=1)
    return df_metric['metric'].sum()/df_metric.shape[0]


print("___Business metric___")
print("LinReg: %.3f" % business_metric(y_test, y_pred))
print("Lasso: %.3f" % business_metric(y_test, y_pred_lasso))
print("ENet: %.3f" % business_metric(y_test, y_pred_enet))
print("Ridge (+name): %.3f" % business_metric(y_test, y_pred_ridge))
```
